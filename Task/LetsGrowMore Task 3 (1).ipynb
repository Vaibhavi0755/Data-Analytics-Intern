```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from PIL import Image

#warnings
import warnings
warnings.filterwarnings("ignore")
```

## Read image

```python
img=cv2.imread("C:\\Users\\JCS\\Downloads\\Shivaji Maharaj.jpg")
```

## Display image in RGB format

```python
image=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis(False)
plt.show()
```

## Converting RGB image to grayscale image

```python
gray_scale_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_scale_img)
```

## imverted image

```python
inverted_image=cv2.bitwise_not(gray_scale_img)
plt.imshow(inverted_image)
```

## Blur image by Gaussian image

```python
blur_image=cv2.GaussianBlur(inverted_image,(111,111), 0)
plt.imshow(blur_image)
```

```python
inverted_blur_img=cv2.bitwise_not(blur_image)
plt.imshow(inverted_blur_img)
```

## Invert the blur image to pencil Sketch image

```python
sketch_img=cv2.divide(gray_scale_img, inverted_blur_img,scale=256.0)
plt.imshow(sketch_img)
```

```python
cv2.imwrite("sketch.png",sketch_img)
```

## Output image

```python
plt.figure(figsize=(8,8))
image1=cv2.cvtColor(sketch_img,cv2.COLOR_RGB2BGR)
plt.imshow(image1)
plt.axis(False)
plt.show()
```
